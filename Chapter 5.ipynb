```python
labels = ['Survived', 'Did not survive']
sizes = [sum(df['Survived'] == 1), sum(df['Survived'] == 0)]
colors = ['lightgreen', 'lightcoral']
explode = (0.1, 0)  # explode the 1st slice (Survived)

# Plotting the pie chart
plt.figure(figsize=(7, 7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.title('Survival Distribution on the Titanic')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame and it has a column named 'Pclass'
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=df, palette='pastel')

# Adding title and labels
plt.title('Count of Passengers by Pclass')
plt.xlabel('Passenger Class')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Counting the number of male and female passengers
gender_counts = df['Sex'].value_counts()

# Plotting a bar chart
plt.figure(figsize=(8, 6))
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'pink'])

# Adding title and labels
plt.title('Number of Male and Female Passengers')
plt.xlabel('Gender')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='purple')

# Adding title and labels
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a grouped DataFrame for the count of survivors by gender
survival_by_gender = df.groupby(['Sex', 'Survived']).size().unstack()

# Plotting a grouped bar chart
survival_by_gender.plot(kind='bar', stacked=False, figsize=(10, 6), color=['lightcoral', 'lightblue'])

# Adding title and labels
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)  # Rotate x-axis labels to be horizontal
plt.legend(['Did not Survive', 'Survived'])

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a grouped DataFrame for the count of survivors by passenger class
survival_by_class = df.groupby(['Pclass', 'Survived']).size().unstack()

# Plotting a grouped bar chart
survival_by_class.plot(kind='bar', stacked=False, figsize=(10, 6), color=['lightcoral', 'lightgreen'])

# Adding title and labels
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)  # Rotate x-axis labels to be horizontal
plt.legend(['Did not Survive', 'Survived'])

# Show the plot
plt.show()
```
